```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import string
%matplotlib inline
train = pd.read_csv('train.csv')
train.head()
```

```python
#Deleting all columns with zero variance.
zer_var_columns = []
all_columns = train.columns
for i in all_columns:
    if train[i].nunique() == 1:
        zer_var_columns.append(i)
```

```python
len(zer_var_columns)
```

```python
train = train.drop(zer_var_columns,axis=1)
```

```python
#Deleting columns with duplicate values.
dup_val_columns = []
all_columns = train.columns
for i in range(len(all_columns)-1):
    v = train[all_columns[i]].values
    dupCols = []
    for j in range(i + 1,len(all_columns)):
        if np.array_equal(v, train[all_columns[j]].values):
            dup_val_columns.append(all_columns[j])
```

```python
train = train.drop(dup_val_columns,axis=1)
```
